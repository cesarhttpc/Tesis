### Diagrama de embudo

```python
import plotly.express as px
data = dict(
    number=[39, 27.4, 20.6, 11, 2],
    stage=["Website visit", "Downloads", "Potential customers", "Requested price", "invoice sent"])
fig = px.funnel(data, x='number', y='stage')
fig.show()
```

### Diagrame de pie concatenado

```python
import plotly.express as px
df = px.data.tips()
fig = px.pie(df, values='total_bill', names='day')
fig.show()
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.sunburst(df, path=['sex', 'day', 'time'], values='total_bill', color='time')
fig.show()
```

```python
import plotly.express as px
import numpy as np
# load data from builtin Plotly data
df = px.data.gapminder()
# prepare a sunburst figure
fig = px.sunburst(df, path=['continent', 'country', 'year'], values='pop',
                color='lifeExp', hover_data=['iso_alpha'],
                color_continuous_scale='twilight',
                color_continuous_midpoint=np.average(df['lifeExp'], weights=df['pop']))
fig.show()
```

### Diagrama (Treemap)

```python
import plotly.express as px
df = px.data.tips()
fig = px.treemap(df, path=['day', 'time', 'sex','smoker'], values='total_bill')
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

### Diagrama 3D puntos

```python

df = px.data.iris()
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_width',
              color='petal_length', size='petal_length', size_max=15,
              symbol='species', opacity=0.7)

# tight layout
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```

### Matriz de diagrama de puntos

```python
df = px.data.iris()
fig = px.scatter_matrix(df,
    dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"],
    color="species", symbol='species')
fig.show()
```

### Diagrama de lineas

```python

df = px.data.iris()
fig = px.parallel_coordinates(df, color="species_id", labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
```

### Diagrama Polar para datos circulares

```python
df = px.data.wind()
fig = px.scatter_polar(df, r="frequency", theta="direction",
                       color="strength", symbol="strength", size="frequency",
                       color_discrete_sequence=px.colors.sequential.Plasma_r)
fig.show()
```

### Mapa

```python

# load built-in gapminder dataset from plotly
gapminder = px.data.gapminder()

# create a map using choropleth
fig = px.choropleth(gapminder, locations='iso_alpha', color='lifeExp', hover_name='country',
                    animation_frame='year', color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show()
```

```python
# Load the dataset
data = px.data.gapminder().query("continent == 'Americas' & year == 2007")

# Create a Choropleth Map
fig = px.choropleth(data_frame=data,
                    locations='iso_alpha',
                    color='pop',
                    hover_name='country',
                    title='Urban Population',
                    color_continuous_scale='blues')

# Set the map layout
fig.update_geos(
    visible=False,
    resolution=50,
    showcountries=True,
    countrycolor="gray",
    showcoastlines=True,
    coastlinecolor="white",
    projection_type="orthographic",  #"natural earth",
    lataxis_range=[10, 40],
    lonaxis_range=[-90, -100]
)

# Display the map
fig.show()
```
